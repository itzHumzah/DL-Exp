```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
from tensorflow.keras import layers

# Load the training dataset
train_data = pd.read_csv("/content/train.csv")

# Load the testing dataset
test_data = pd.read_csv("/content/test.csv")

# Extract features and target variable from training dataset
X_train = train_data.drop("price_range", axis=1)
y_train = train_data["price_range"]

# Extract features from testing dataset and remove 'id' column
X_test = test_data.drop("id", axis=1)

# Normalize the features in both training and testing datasets
scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)

# One-hot encode the target variable in the training dataset
y_train_encoded = keras.utils.to_categorical(y_train, num_classes=4)  # Assuming there are 4 price ranges (0 to 3)

# Define the Neural Network model using Keras
model = keras.Sequential([
    layers.Dense(16, activation='relu', input_dim=20),
    layers.Dense(12, activation='relu'),
    layers.Dense(4, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()

# Train the model on the training data
model.fit(X_train_normalized, y_train_encoded, epochs=50, batch_size=32, validation_split=0.1)

# Evaluate the model on the test data
y_pred = model.predict(X_test_normalized)

# Convert the one-hot encoded predictions back to class labels (0 to 3)
predicted_labels = np.argmax(y_pred, axis=1)
```
